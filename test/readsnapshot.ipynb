```python
import h5py
import numpy as np
import sys
sys.path.append('../')
from KunRead.io_utils import * 

### index of cosmology from 0 to 128
ic    = 0
### index of snapshot file from 0 to 11 --> z [3.0, 0.0]
isnap = 0
### main directory
fdir  = '/home/cossim/yuyu22/chenzhao/csst/simulation/'
fbase = fdir + 'c%04d/output/snapdir_%03d_down64/snapshot_%03d_down64'%(ic, isnap, isnap)
##### Basic information
with h5py.File(fbase + '.0.hdf5', 'r') as ff:
    print(ff.keys())
    print(ff['Header'].attrs.keys())
    NumFiles = ff['Header'].attrs['NumFilesPerSnapshot']
    NumPartTot = ff['Header'].attrs['NumPart_Total'][1] ## 1 means Dark matter
    print(NumFiles, NumPartTot)
    print(ff['PartType1'].keys())
    print(ff['PartType1/Coordinates'].attrs.keys())
```

```python
fname = fbase + '.0.hdf5'  # Example file name for single snapshot file

# Example 1: Single file reading
print("Example 1: Single snapshot file reading")
data = IO_Snap_single(
    fname,
    return_pos=True,
    return_vel=True,
    return_id=True,
    part_type=1,
    dtype=np.float32,
    id_dtype=np.int64
)

if data is not None:
    print(f"Read {len(data)} particles from single file")
    print(f"Data type: {data.dtype}")
    print(f"Available fields: {data.dtype.names}")
    if 'Pos' in data.dtype.names:
        print(f"Position shape: {data['Pos'].shape}")

# Example 2: Get snapshot information
print("\nExample 2: Snapshot information")
info = get_snapshot_info(fname)
print(f"Number of files: {info.get('NumFiles', 1)}")
print(f"Total particles: {info.get('NumPart_Total', [0,0,0,0,0,0])}")
print(f"Available particle types: {info.get('AvailablePartTypes', [])}")

# Example 3: Multi-file reading with multiprocessing
print("\nExample 3: Multi-file snapshot reading")
all_particles = IO_Snap_multifile(
    fname,
    return_pos=True,
    return_vel=True,
    return_id=True,
    part_type=1,
    dtype=np.float32,
    id_dtype=np.int64,
    n_processes=4
)

if all_particles is not None:
    print(f"Read {len(all_particles)} particles from all files")

# Example 4: Get total particle count
print("\nExample 4: Total particle count")
total_count = get_total_particle_count(fname, part_type=1)
print(f"Total particles of type 1: {total_count}")
```
