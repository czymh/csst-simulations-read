```python
print(single_file_data.dtype)
def AddPosFromXYZ(data):
    from numpy.lib.recfunctions import drop_fields, repack_fields
    org_dtype = data.dtype.descr
    new_dtype = data.dtype.descr + [('Pos', '<f8', (3,))]
    new_data  = np.zeros(data.shape, dtype=new_dtype)
    for name in data.dtype.names:
        new_data[name] = data[name]
    new_data['Pos'][:, 0] = data['X']
    new_data['Pos'][:, 1] = data['Y']
    new_data['Pos'][:, 2] = data['Z']
    new_data = drop_fields(new_data, ["X", "Y", "Z"])
    new_data = repack_fields(new_data)
    return new_data
data = AddPosFromXYZ(single_file_data)
print(data.dtype)
```
